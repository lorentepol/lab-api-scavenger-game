## Challenge 3: Hidden Cold Joke

Using Python, call Github API to find out the cold joke contained in the 24 secret files in the following repo:

https://github.com/ironhack-datalabs/scavenger

The filenames of the secret files contain .scavengerhunt and they are scattered in different directories of this repo. The secret files are named from .0001.scavengerhunt to .0024.scavengerhunt. They are scattered randomly throughout this repo. You need to search for these files by calling the Github API, not searching the local files on your computer.

#### 1. Find the secret files.

`````python
url = "https://api.github.com/repos/ironhack-datalabs/scavenger/contents/"

payload = {}
headers= {}

response_scavenger = requests.request("GET", url, headers=headers, data = payload)
`````

`````python
json_response_scavenger = response_scavenger.json()
df_scavenger_folders = pd.json_normalize(json_response_scavenger)
list_folders = df_scavenger_folders[df_scavenger_folders['type'] == 'dir']['name'].to_list()
list_folders
`````

`````python
import getpass
# To keep your token secured
username = 'lorentepol'
token = getpass.getpass(prompt='Pol, please enter your token here:')
`````

`````python
all_files_scavenger = []
for folder in list_folders:
    url = "https://api.github.com/repos/ironhack-datalabs/scavenger/contents/" + folder
    files_scavenger = requests.request("GET", url, auth=(username, token), headers=headers, data = payload)
    all_files_scavenger += files_scavenger.json()
`````

`````python
df_all_files = pd.json_normalize(all_files_scavenger)
`````

`````python
df_files_scavenger = df_all_files[df_all_files['name'].str.contains('\.\d{4}\.scavengerhunt')]
df_files_scavenger
`````

#### 2.  Sort the filenames ascendingly.

`````python
df_files_scavenger = df_files_scavenger.sort_values('name')

list_paths_scavenger = df_files_scavenger['path'].to_list()
list_paths_scavenger
`````

#### 3. Read the content of each secret files into an array of strings.
Since the response is encoded, you will need to send the following information in the header of your request:
````python
headers = {'Accept': 'application/vnd.github.v3.raw'}
````

`````python
all_content_scavenger = []
for path in list_paths_scavenger:
    url = "https://api.github.com/repos/ironhack-datalabs/scavenger/contents/" + path
    headers = {'Accept': 'application/vnd.github.v3.raw'}
    content_scavenger = requests.request("GET", url, auth=(username, token), headers=headers)
    all_content_scavenger += content_scavenger.text
`````

`````python
all_content_scavenger
`````

#### 4. Concatenate the strings in the array separating each two with a whitespace.

`````python
joke = ''.join(all_content_scavenger)
`````

#### 5. Print out the joke.

`````python
print(joke)
`````
